```python
# This is the brain of  my todo list
def main():
  listify()
  choose_task()
```

```python
# we store our menu in a list
activity = ["Add Task", "Task Complete", "List Task", "Remove Task", "Quit"]
# This function creates a menu for our todo list
def actions():
  for index, act in enumerate(activity):
    print(f"      {index + 1}: {act}")
```

```python
# Display msg on home screen
def listify():
  print()
  print("WELCOME TO YOUR TO-DO LIST MENU")
  print()
  actions()
  print()
```

```python
# Navigate the menu by matching the numbering
# done
def choose_task():
  menu = int(input("Enter the number matching the menu item you want to navigate "))
  if menu == 1:
    print()
    lines = int(input("Enter the number of tasks you have "))
    add_task(lines)
  elif menu == 2:
    print()
    mark_complete()
  elif menu == 3:
    print()
    my_list_of_task()
  elif menu == 4:
    print()
    remove_task()
  elif menu == 5:
    print()
    quit_todo()
  else:
    print()
    print("Invalid Task Number")
```

```python
# here the user has decided to terminate task list

# DONE

def quit_todo():
  msg = "goodBye, We hope you visit again"
  print(msg.upper())
  print("to return to the main menu, you can copy < main() > in a new cell and run(shift+enter)")
```

```python
# User is able to add his todos by stating the number of todos they want to note down.
# if number of todos is attained, the list displays a msg full.
# the todos are stored in a list(list_of_todos)
# This fuction is called inside the <choose_task function>

# DONE

list_of_todos = []
def add_task(no_of_todo):
  todo = 0
  print("Task_Description ")
  while True:
    if todo < no_of_todo:
      list_task = input()
      # list_of_todos.append(list_task.capitalize())
      list_of_todos.append(list_task.capitalize())
      todo += 1
    else:
      add_message = "Task list is full"
      print(add_message.capitalize())
      break
  return list_of_todos

```

```python
# list of numbered items in my todo list
def my_list():
  for index, item in enumerate(list_of_todos):
    print(f"{index+1}: {item}")
```

```python
def task_complete():
  for index, taskList in enumerate(fully_completed):
    print(f"{index+1}: {taskList}")
```

```python
fully_completed = []
complete_index = [] #holds the index of a completed task

def mark_complete():
  my_list() #helps user view listing to identify as complete
  print("")
  if len(list_of_todos) > 0:
    completed_task = int(input("Enter the number matching the task to mark as complete "))
    for index, listings in enumerate(list_of_todos):
      if completed_task == index+1:
        fully_completed.append(listings)
        complete_index.append(index)
        print(fully_completed, complete_index)
        print("")
    print("The tasks below have been completed")
    task_complete()
    print("You can proceed and delete them by entering 4")
    main()
  else:
    empty_todo()
```

```python
# iterates over our list where we appended our added tasks
# prints out numbered task while also displaying if completed
def my_list_of_task():
  if len(list_of_todos) > 0:
    print("This is my TO-DO list")
    for index, listings in enumerate(list_of_todos):
      # for list_complete_index in complete_index:
        if index in complete_index:
          print(f" {index+1}: {listings} - COMPLETE")
        else:
          print(f" {index+1}: {listings} - INCOMPLETE")
    main()
  else:
    empty_todo()
```

```python
def empty_todo():
  print("You have an empty TO_DO list")
  print("Kindly enter 1 to add your Tasks")
  print("")
  main()
```

```python
# delete = ["Delete All Tasks", "Delete Completed Tasks Only", "Delete By Number"]
def remove_task():
    if len(list_of_todos) > 0:
      my_list() # Displays your list items in the to-do list
      num_to_delete = int(input("Enter the number to delete from the list above"))
      for index, delete in enumerate(list_of_todos):
        if index+1 == num_to_delete:
          print("---***---")
          print(f"{delete} has been deleted")
          print("---***---")
          break
      list_of_todos.remove(delete)
      main()
    else:
      empty_todo()
```
